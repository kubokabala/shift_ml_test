# **Задание 10**
Возьмем две группы заемщиков с рейтингом = 'Ж". Первая группа займ выплатила (итоговый_статус_займа = 1), вторая группа нет (итоговый_статус_займа = 0). Аналитиков интересует, является ли различие в средней сумме займа у этих групп статистически значимым при a=0.05 (использовать t-критерий для независимых выборок):?


Выделим группы

```python
from scipy.stats import ttest_ind

df_loan1 = df[(df['рейтинг'] == 'Ж') & (df['итоговый_статус_займа'] == 1)]['сумма_займа']
df_loan2 = df[(df['рейтинг'] == 'Ж') & (df['итоговый_статус_займа'] == 0)]['сумма_займа']
df_loan1
```

Проверим, сопоставимы ли они по размеру

```python
len(df_loan1), len(df_loan2)
```

Да. сопоставимы. Итак, нулевая гипотеза следующая:

_Различие не является статистически значимым_, т.е. p-value меньше уровня значимости.

```python
t_stat, p_value = ttest_ind(df_loan1, df_loan2, equal_var=True)
print(f"t-статистика: {t_stat}")
print(f"p-значение: {p_value}")
```

p > 0.05 => разница не является статистически значимой, нулевая гипотеза подтверждается.
